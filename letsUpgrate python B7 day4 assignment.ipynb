```python
# print the first Armstrong number in the given rangeof 1042000 to 702648265 
# and exit loop as soon as you encounter first Armstrong no.
```

```python
for n in range (1042000,702648265):
    l=len(str(n))
    sum =0
    temp=n
    
    while temp>0:
        d=temp%10
        sum+=d**l
        temp//=10
    if n==sum:
        print(n)
        break
```
